```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = pd.read_csv('/content/tested.csv')
data.head(10)
```

```python
age_mean = data['Age'].mean()
age_median = data['Age'].median()
age_mode = data['Age'].mode()
print("mean of the age:",age_mean)
print("median of the age:",age_median)
print("mode of the age:",age_mode)
```

```python
fare_mean = data['Fare'].mean()
fare_median = data['Fare'].median()
fare_mode = data['Fare'].mode()
print("mean of the age:",fare_mean)
print("median of the age:",fare_median)
print("mode of the age:",fare_mode)
```

```python

data['Survived'] = data['Survived'].replace({1:'Survived',0:'Not survived'})
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.countplot(x='Survived',hue = 'Pclass',data=data)
plt.title('Survived and Pclass')
plt.xlabel('survive')
plt.ylabel('Pclass')
plt.show()
```

```python
null_vlues = data.isnull().sum()
print('null values in each column')
print(null_vlues)
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Fare'] = data['Fare'].fillna(data['Fare'].mode())
print ("\nNull values after column")
print(data.isnull().sum())
```
